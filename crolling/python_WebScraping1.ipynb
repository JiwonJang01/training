#할리스 웹사이트 매장찾기 페이지 크롤링

```python
from bs4 import BeautifulSoup         # 소스를 파싱하는 라이브러리
import urllib.request as ur           # url을 열기 위한 확장라이브러리
```

```python
# 작업결과를 저장할 리스트 준비
result = []
```

```python
hollys_url = 'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=1&sido=&gugun=&store='   # 할리스 홈페이지 매장지도 url
html = ur.urlopen(hollys_url)     # url을 열어 html에 저장한다.

# 객체 생성
Soupholly = BeautifulSoup(html,'html.parser')     # html에 저장된 컨텐츠를 html parser를 이용해서 이해하고 Soupholly에 저장한다.
print(Soupholly.prettify())       # Soupholly에 저장된 내용 중 필요한 내용이 들어있는지 확인한다.
```

```python
# 해당 사이트에서 개발자 모드를 열어 필요한 영역의 태그를 확인하고 카피한다.
# 객체에서 select를 사용하여 필요한 영역의 태그 부분만 따로 저장한다
stores = Soupholly.select('#contents > div.content > fieldset > fieldset > div.tableType01 > table > tbody > tr')
print(stores[0])
store_td = stores[0].find_all('td')         # 해당 영역의 첫번째 인덱스에서 td태그를 가진 모든 부분을 store_td에 따로 저장 해주고
print(store_td)                 # store_td를 출력하여 확인해준다
```

```python
stores = Soupholly.select('#contents > div.content > fieldset > fieldset > div.tableType01 > table > tbody > tr')
                            # 영역태그 부분을 매장찾기의 2번째 페이지에서 카피해왔다. 같은 영역이지만 다른 내용이 들어있다.
# 카피해 온 컨텐츠에서 td태그 부분을 찾아 string형식인 부분을 store_list에 차례대로 추가되게 해준다,
for store in stores:
  store_list = []
  for i in store.find_all('td'):
    store_list.append(i.string)
result.append(store_list)
print(result)
```

```python
# 위에 두 코드를 실행시키면 공통적으로 None으로 표시되는 부분이 있다
# 웹사이트에서는 이미지로 표시되는 부분이다. 이미지로 표시되지만 alt태그를 보면 표시 이름이 적혀있다.
stores[0].find_all('td')[4].find_all('img')[0].attrs['alt']
```

```python
result = []
for store in stores:  # 페이지내 매장정보를 가지고 있는 TR 태그들의 모음
  store_list = []
  for idx,i in enumerate(store.find_all('td')): # TR에는 각 컬럼에 해당하는 TD 태그들의 모음
    if idx == 4:      # 인덱스가 4 즉, 이미지 표시가 있는 부분
      temp = []
      for i4 in i.find_all('img'):  # 5번째(인덱스는 4) 컬럼은 텍스트 정보가 없고 이미지 정보가 있음
        temp.append(i4.attrs['alt'])  # 이미지 정보를 temp에 추가해줌
      store_list.append(','.join(temp))   # temp리스트와 store_list를 조인시켜준다.
    else:
      store_list.append(i.string)
  result.append(store_list)   # 이미지 표시까지 포함되어 완성된 리스트들을 result에 추가시켜준다.
```

```python
# 위의 과정을 함수로 만들어서 함수를 페이지 번호를 1~51까지 실행
def getHollyStore(pageNum):
  hollys_url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={pageNum}&sido=&gugun=&store='
  html = ur.urlopen(hollys_url)
  Soupholly = BeautifulSoup(html,'html.parser')
  stores = Soupholly.select('#contents > div.content > fieldset > fieldset > div.tableType01 > table > tbody > tr')
  for store in stores:  # 페이지내 매장정보를 가지고 있는 TR 태그들의 모음
    store_list = []
  for idx,i in enumerate(store.find_all('td')): # TR에는 각 컬럼에 해당하는 TD 태그들의 모음
    if idx == 4:
      temp = []
      for i4 in i.find_all('img'):  # 5번째(인덱스는 4) 컬럼은 텍스트 정보가 없고 이미지 정보가 있음
        temp.append(i4.attrs['alt'])
      store_list.append(','.join(temp))
    else:
      store_list.append(i.string)
  result.append(store_list)
  return result
```

```python
result = []
for pageNum in range(5,10):
  result += getHollyStore(pageNum)
```

```python
import pandas as pd
df = pd.DataFrame(result,columns=['지역', '매장명', '현황', '주소', '매장서비스', '전화번호'])    # 데이터프레임을 만들어준다
df.head(10)
```
