# 하나의 이미지 추출

```python
  import requests     # 응답객체
  import os           # join메서드 활용 새 경로 생성
```

```python
# 이미지 추출 함수로 만들기
def saveImage(url):
  html_image = requests.get(url)
  img_file_name = os.path.basename(url)
  with open('./img/'+img_file_name, 'wb') as f:   # img라는 폴더가 있어야함
    chunk_size = 1000000
    for chunk in  html_image.iter_content(chunk_size):
      f.write(chunk)
```

```python
url = 'https://img.hankyung.com/photo/202308/PYH2023082410230006100_P4.jpg'
saveImage(url)
```

# 이미지가 있는 사이트에서 이미지 추출

```python
# 이미지가 있는 사이트에서 이미지 추출
import urllib.request as ur
from bs4 import BeautifulSoup

# 이미지 파일명을 수집하여 정보확인
site_url = 'https://news.naver.com/'
html = ur.urlopen(site_url)
soup = BeautifulSoup(html,'html.parser')
print(soup.prettify())    # 수집할 정보가 html형식으로 가져왔는지 확인
```

```python
for img in soup.find_all('img'):
  if 'src' in img.attrs.keys(): # 딕셔너리 구조를 기억할 것
    url = img.attrs['src'].split("?")[0]    # src 태그를 가진 이미지들을 가져오기
    saveImage(url)
```
