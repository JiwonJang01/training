```python
from ultralytics import YOLO
import cv2

model = YOLO('yolov8s.pt')
```

```python
result = model.predict('/content/bb.jpg', save=True,conf = 0.8)
```

```python
detect_img = result[0].save_dir+'/'+'bb.jpg'
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,12))
img = cv2.imread(detect_img)
plt.imshow(img)
```
