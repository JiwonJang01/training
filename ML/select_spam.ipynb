```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/kyuyounglee/data/main/spam.csv', encoding='utf-8')
df
```

```python
df.describe()
```

```python
# 전처리  : 특수기호 제거
import string
string.punctuation
```

```python
# 문자열 -> 문자를 하나씩 꺼내서 특수문자를 제거하고 다시 문자열로 합친다 -> 판다스형태로 저장
```

```python
"".join(v for v in df.text[0] if v not in string.punctuation)
```

```python
df['text'] = df.text.apply(lambda row : "".join(v for v in row if v not in string.punctuation) )
```

```python
df
```

```python
# 불용어 제거  stopword
```

```python
pip install nltk
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
stopwords.fileids()  # 라이브러리가 제공하는 불용어 언어들  24개정도
```

```python
df.text =  df.text.apply(
    lambda row : 
        ' '.join(v for v in row.split() if v not in stopwords.words('english') ).lower() 
             )
```

```python
df
```

```python
# sapm:1 ham:0
```

```python
df.target = df.target.map(
    {'spam':1, 'ham':0}
)
```

```python
df.head()
```

```python
# 자연어처리
# 카운트 기반 벡터 : 문자를 개수기반으로 벡터화  하고 문장을 숫자로 표현
# 각 문장마다 등장한 단어의 수를 센다  (  (행번호,인덱스번호), 출현횟수 )
```

```python
X = df.text
y = df.target
```

```python
pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
cv.fit(X)
cv.vocabulary_  # 단어와 인덱스를 출력 BOW
```

```python
x = cv.transform(X)
print(x)
```

```python
df.text[0]
```

```python
cv.vocabulary_['go']
```

```python
x,y
```

```python
# 모델링 예측 / 평가
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(x, y, stratify = y, test_size = 0.2, random_state=42)
```

```python
# 나이브 베이지 모델 - 모든 특징들은 독립적이라는 가정 하에
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
accuracy_score(y_test,pred)
```

```python
print(confusion_matrix(y_test,pred))
```

```python
model.classes_
```

```python
print(classification_report(y_test, pred))
```

```python
roc_auc_score(y_test, pred)
```
