```python
import numpy as np
import tensorflow as tf
from PIL import Image
from glob import glob
import os
```

```python
# 자신의 구굴 드라이브 내 모델 위치
# 자신의 구글 드라이브내의 모델 위치
cnn_model = tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/best_model.keras')
cnn_model
```

```python
target_name = [
'airplane'                    ,
'automobile'                  ,
'bird'                    ,
'cat'                   ,
'deer'                  ,
'dog'                   ,
'frog'                  ,
'horse'                 ,
'ship'                  ,
'truck',
]
```

```python
os.listdir('/content/test_images')
```

```python
files_list = glob('/content/test_images/*.jpg')
file_names = [path.split('/')[-1] for path in files_list]
```

```python
# 이미지를 읽어서 - 데이터 ... 사이즈는 학습에 입력으로 들어가 사이즈와 맞춰야 한다.
# CFIR-10 (32,32)
img = Image.open(files_list[0])
np.array(img.resize( (32,32) )) / 255.0
```

```python
X_train = []
for filename in files_list :
  img = Image.open(filename)
  X_train.append( np.array(img.resize((32,32)) ) / 25.0)
```

```python
X_train = np.array(X_train)
X_train.shape   #(12,32,32,3)
```

```python
# 예측
preds = cnn_model.predict(X_train)
```

```python
preds[0]
```

```python
# preds의 가장 큰 값을 가진 인덱스를 알아야함
preds[0], np.argmax(preds[0]), target_name[np.argmax(preds[0])]
```

```python
[target_name[np.argmax(p)] for p in preds]
```

```python
if not os.path.isdir('predict'):
  os.mkdir('predict')
  # 해당 위치로 이동
os.chdir('./predict')
```

```python
# 분류별로 폴더 만들기
for name in target_name:
  if not os.path.isdir(name):
    os.mkdir(name)
```

```python
# 예측한 이름에 해당하는 폴더로 해당 이미지 복사(이동)
# 예측한 이름
import shutil     # 복사 라이브러리
os.chdir('/content/predict')

predict_names = [target_name[np.argmax(p)]for p in preds]
for name,file in zip(predict_names,files_list) :
  target_path = '/content/predict/'+name+'/'
  file_name = file.split('/')[-1]
  os.chdir(name)    # 해당 폴더로 이동
  shutil.copyfile(file,target_path+file_name)       # 오리지널 파일을 해당 폴더의 오리지널 파일명으로 복사
  os.chdir('../')     # 다시 한단계 위로 이
```
