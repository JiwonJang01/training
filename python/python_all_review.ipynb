```python
# 철수가 상점에 가서 3,200원짜리 물건을 샀다
# 그런대 돈을 10,000원을 지불했다
# 얼마를 거슬러 줘야 하나
# 출력예)5000:1, 1000:1, 500:1, 100:3

# 조건문과 나머지 연산을 이용해서 풀어보자

pay = int(input("지불한 금액을 입력하세요 : "))
buy = int(input("물건의 금액을 입력하세요 : "))
cashback = pay - buy

while cashback > 0:
    if cashback >= 5000:
        count = cashback // 5000
        cashback = cashback - 5000 * count
        print(f"5000원 : {count}")
    elif cashback >= 1000:
        count = cashback // 1000
        cashback = cashback - 1000 * count
        print(f"1000 원: {count}")
    elif cashback >= 500:
        count = cashback // 500
        cashback = cashback - 500 * count
        print(f"500원 : {count}")
    elif cashback >= 100:
        count = cashback // 100
        cashback = cashback - 100 * count
        print(f"100원 : {count}")
    else:
      print("거스름돈이 없습니다.")
```

```python
# 다음 문장에서 가장 긴 문자열을 찾아서 출력
text = '동해물과 백두산이 마르고 닳고록 하느님이 보우하사 우리나라만세'
# 공백을 기준으로 분리하여 리스트에 담기
wors = text.split()
maxlength = 0
maxWord = ''
# 리스트를 순환하며 가장 긴 문장을 찾기
for txt in wors:
  if maxlength < len(txt):
    maxlength = len(txt)
    maxWord = txt
print(f'가장 길이가 긴 문자열은 : {maxWord}')
```

```python
# 리스트.sort() 오름차순으로 정렬
# 주어진 숫자들의 집합(리스트)에서 중복을 제거하고 내림차순으로 정렬
import random
numbers = random.sample(range(20),10)
# numbers의 중복을 제거하여 리스트로 만들기
numbers = list(set(numbers))
numbers.sort()
# 리스트 슬라이싱을 통해 내림차순 정렬
print(numbers)
print(numbers[::-1])
```

```python
# 함수를 만들고 호출: 매개변수 존재, 리턴값 존재
# 함수를 호출한 후에 결과값을 받아서 출력하는
# 임의의 데이터를 갖는 리스트를 받아서 짝수의 개수를 리턴하는 함수
####from~import count
import random
def countEven(lists):
  return sum([i for i in lists if i%2 == 0]) # 리스트 컴프리핸션

myLists = random.sample(range(100),20)
print(myLists)
evenCount = countEven(myLists)
print(evenCount)
```
